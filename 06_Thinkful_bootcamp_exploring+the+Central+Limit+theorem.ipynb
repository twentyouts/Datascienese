```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
pop1 = np.random.binomial(10,0.2,10000)
pop2 = np.random.binomial(10,0.5,10000)

#histogramm of two grouprs

plt.hist(pop1,alpha = 0.5, label = 'Population 1')
plt.hist(pop2,alpha = 0.5, label = 'Population 2')
plt.legend(loc = 'upper right')
plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace = True)
sample2 = np.random.choice(pop2, 100, replace = True)

plt.hist(sample1,alpha = 0.5, label = 'Sample 1')
plt.hist(sample2,alpha = 0.5, label = 'Sample 2')
plt.legend(loc = 'upper right')
plt.show()
```

```python
print(pop1.mean())
print(pop2.mean())
print(pop1.std())
print(pop2.std())


print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
#increasing sample from 100 to 1000 i expect to see descriptive statistic closer to the population distribution. 
sample3 = np.random.choice(pop1, 1000, replace = True)
sample4 = np.random.choice(pop2, 1000, replace = True)

plt.hist(sample3,alpha = 0.5, label = 'Sample 1')
plt.hist(sample4,alpha = 0.5, label = 'Sample 2')
plt.legend(loc = 'upper right')
plt.show()

print(sample3.mean())
print(sample4.mean())
print(sample3.std())
print(sample4.std())
```

```python
#so if we decrease sample size  descriptive statistic would be farther from population distribution. It's values will be more 
#random each time we choose 20 sample
sample5 = np.random.choice(pop1, 20, replace = True)
sample6 = np.random.choice(pop2, 20, replace = True)

plt.hist(sample5,alpha = 0.5, label = 'Sample 1')
plt.hist(sample6,alpha = 0.5, label = 'Sample 2')
plt.legend(loc = 'upper right')
plt.show()


print(pop1.mean())
print(pop2.mean())
print(pop1.std())
print(pop2.std())


print(sample5.mean())
print(sample6.mean())
print(sample5.std())
print(sample6.std())
```

```python
#when we change p value for pop3 from 0.2 to 0.3 means of two distributions will be closer
pop3 = np.random.binomial(10,0.3,10000)
pop4 = np.random.binomial(10,0.5,10000)

#histogramm of two grouprs

plt.hist(pop1,alpha = 0.5, label = 'Population 1')
plt.hist(pop2,alpha = 0.5, label = 'Population 2')
plt.legend(loc = 'upper right')
plt.show()

print(pop3.mean())
print(pop4.mean())
print(pop3.std())
print(pop4.std())
```

```python
#when we change p value for pop3 from 0.4 to 0.4 means of two distributions will be even more closer. What changed? The overlap
#area of two distributions increased while their means moving towards
pop3 = np.random.binomial(10,0.4,10000)
pop4 = np.random.binomial(10,0.5,10000)

#histogramm of two grouprs

plt.hist(pop1,alpha = 0.5, label = 'Population 1')
plt.hist(pop2,alpha = 0.5, label = 'Population 2')
plt.legend(loc = 'upper right')
plt.show()

print(pop3.mean())
print(pop4.mean())
print(pop3.std())
print(pop4.std())
```
