```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pareto = np.random.pareto(2, 100)
plt.hist(pareto,color = 'b', alpha = 0.5)
plt.title('Pareto distribution')
plt.axvline(pareto.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(pareto.mean() + pareto.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean() - pareto.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
gamma = np.random.gamma(5,100, 100)
plt.hist(gamma,color = 'b', alpha = 0.5)
plt.title('Gamma distribution')
plt.axvline(gamma.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
lognormal = np.random.lognormal(3,1, 100)
plt.hist(lognormal,color = 'b', alpha = 0.5)
plt.title('lognormal')
plt.axvline(lognormal.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(lognormal.mean() + lognormal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(lognormal.mean() - lognormal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
logistic = np.random.logistic(3,1, 100)
plt.hist(logistic,color = 'b', alpha = 0.5)
plt.title('lognormal')
plt.axvline(logistic.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(logistic.mean() + logistic.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(logistic.mean() - logistic.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
power = np.random.power(5, 100)
plt.hist(power,color = 'b', alpha = 0.5)
plt.title('power')
plt.axvline(power.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(power.mean() + power.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(power.mean() - power.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
triangular = np.random.triangular(-5, 2 ,8, 100000)
plt.hist(triangular,bins = 200,color = 'b', alpha = 0.5)
plt.title('triangular')
plt.axvline(triangular.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(triangular.mean() + triangular.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(triangular.mean() - triangular.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
normal_var1 = np.random.normal(5,0.5,100000)
normal_var2 = np.random.normal(10,1,100000)
normal_var3 = normal_var1 + normal_var2 
```

```python
plt.hist(normal_var3,bins = 200,color = 'b', alpha = 0.5)
plt.title('normal_var3')
plt.axvline(normal_var3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normal_var3.mean() + normal_var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normal_var3.mean() - normal_var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

print(normal_var3.std())
print(normal_var3.mean())
```
